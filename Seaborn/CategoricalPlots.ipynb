# 1) Categorical scatter plot

<p><h4 style="display:inline"> Stripplot </h4>: It is a axis level function.</p>
 It is particularly useful for visualizing the distribution and spread of data points when dealing with categorical or numerical data.

```python
sns.stripplot(data=d1,x="day",y="total_bill")
```

The jitter parameter controls the magnitude of jitter or disables it altogether.

```python
sns.stripplot(data=d1,x="day",y="total_bill",jitter=False)
```

<p><h4 style="display:inline">catplot</h4>: It is a figure level plot</p>

```python
sns.catplot(data=d1,x="sex",y="total_bill",kind="strip")
```

```python
sns.catplot(data=d1,x="day",y="total_bill",kind="strip",jitter=.2,hue="sex")
```

<p><h4 style="display:inline">swarmplot</h4>: It is a axis level plot</p>

```python
sns.swarmplot(data=d1,x="day",y="tip",hue="smoker")
```

```python
sns.catplot(data=d1,x="day",y="tip",hue="sex")
```

# 2) Categorical distribution plot.

<p><h4 style="display:inline">Box Plot</h4>:
A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the data's summary statistics, such as the median, quartiles, and potential outliers

```python
sns.catplot(data=d1,x="sex",y="total_bill",kind="box")
```

```python
sns.catplot(data=d1,x="day",y="total_bill",kind="box")
```

```python
sns.catplot(data=d1,x="day",y="total_bill",hue="sex",kind="box")
```

```python
sns.catplot(data=d1,y="tip",kind="box")
```

<p><h4 style="display:inline">Boxenplot</h4>: It is used to create a specialized type of box plot, known as a "letter-value box plot" or "letter-value plot". This plot is designed to provide a more detailed view of the distribution of data compared to a traditional box plot. It is best suited for larger datasets.</p>


```python
d3=pd.read_csv("D:\Python\Datasets\diamond.csv")
d3
```

```python
sns.catplot(data=d3,x="color",y="price",kind="box")
```

```python
sns.catplot(data=d3.sort_values("color"),x="color",y="price",kind="boxen")
```

```python
sns.catplot(data=d3,x="cut",y="price",kind="boxen")
```

<p><h4 style="display:inline">Violinplot:</h4>
A violin plot is a data visualization that combines elements of a box plot with a kernel density plot. It is used to visualize the distribution of numerical data across different categories or groups.</p>

```python
sns.violinplot(data=d1,x="day",y='total_bill')
```

```python
sns.catplot(data=d1,x="day",y="total_bill",kind="violin")
```

```python
sns.catplot(data=d1,x="day",y="total_bill",kind="violin",hue="sex")
```

```python
sns.catplot(data=d1,x="day",y="total_bill",kind="violin",hue="sex",split=True)
```

# 3) Categorical estimate plot

<p><h4 style="display:inline">Bar Plot:</h4> In a bar plot, the length or height of each bar corresponds to the aggregate value or count associated with a particular category.<p>

```python
sns.barplot(data=d1,x="sex",y="total_bill",estimator="mean",ci=None)
```

```python
sns.barplot(data=d1,x="sex",y="total_bill",estimator="mean")
```

```python
sns.barplot(data=d1,x="sex",y="total_bill",estimator="mean",hue="smoker")
```

```python
import numpy as np
sns.barplot(data=d1,x="sex",y="total_bill",estimator=np.std)
```

```python
sns.barplot(data=d1,x="sex",y="total_bill",estimator=np.min,hue="smoker")
```

```python
sns.barplot(data=d1,x="sex",y="total_bill",estimator=np.max,hue="smoker")
```

<p><h4 style="display:inline">Point Plot:</h4>  Point plots are often used to show trends, patterns, or comparisons between different categories or groups. They are particularly useful for visualizing how a particular value changes across categories.<p>

```python
sns.pointplot(data=d1,x="sex",y="total_bill")
```

```python
sns.pointplot(data=d1,x="sex",y="total_bill",hue="smoker")
```

```python
# Countplot
sns.countplot(data=d1,x="sex",hue="day")
```
