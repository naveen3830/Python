# Regression plot

* regplot - axes level function
* lmplot - figure level function

In the simplest invocation both functions draw a scatterplot of two variables, x and y, and then fit the regression model y~x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
sns.regplot(data=d1,x="total_bill",y="tip")
```

```python
sns.lmplot(data=d1,x="total_bill",y="tip",hue="sex")
```

<p><h4 style="display:inline">Residplot: </h4>Residplot is a function in the Seaborn data visualization library that is used to plot the residuals of a linear regression. Residuals are the differences between the observed and predicted values in a regression analysis. This plot helps you assess the goodness of fit of a regression model and identify patterns in the residuals.</p>

```python
sns.residplot(data=d1,x="total_bill",y="tip")
```

# Plotting Pairwise Relationsship

```python
import pandas as pd
d2=pd.read_csv("D:/Python/Datasets/iris.csv")
d2
```

<p><h4 style="display:inline"> Pairplot: </h4>Pairplot is a function in the Seaborn data visualization library that creates a grid of scatterplots for each pair of numerical features in a dataset. It also includes histograms along the diagonal to show the distribution of each individual variable. This plot is a great way to quickly visualize relationships between variables in a dataset.</p>

```python
sns.pairplot(d2)
```

```python
sns.pairplot(d2,hue="Species")
```

# Pair grid

```python
g=sns.PairGrid(data=d2)
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=d2)
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(data=d1,hue="sex")
g.map_diag(sns.stripplot)
g.map_offdiag(sns.kdeplot)
```

```python
g=sns.PairGrid(data=d1,hue="sex")
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
d2
```

```python
g=sns.PairGrid(data=d2,hue="Species",vars=["SepalWidthCm","SepalLengthCm"])
g.map_diag(sns.stripplot)
g.map_offdiag(sns.kdeplot)
```
