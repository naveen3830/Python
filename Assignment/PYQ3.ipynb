```python
bins = [0, 20, 30,40 ,50,float('inf')]
labels = ['<20 years', '20-30 years', '30-40 years',"40-50 years",">50 years"]

df['age_bucketed'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
department_mapping = {
    'HR': 'HR001',
    'Sales': 'SAL002',
    'Manager': 'MGR003',
    'Developer': 'DEV004'
}
```

```python
df['Function_Code'] = df['Function'].map(department_mapping)
df.sample(10)
```

```python
Q1_val = df['Salary'].quantile(0.25)
Q3_val = df['Salary'].quantile(0.75)
IQR = Q3_val - Q1_val
lower_bound = Q1_val - 1.5 * IQR
upper_bound = Q3_val + 1.5 * IQR
df['Outlier_Flag'] = np.where((df['Salary'] < lower_bound) | (df['Salary'] > upper_bound),
                            "Outlier", "Not an Outlier")

df.sample(10)
```
