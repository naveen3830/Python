# Merge the three data sets  with sales, TV GRPs & Video Impressions

```python
# Importing the datasets
sales_data = pd.read_csv(r'D:\Christ\Python\Assignment\sales_data.csv')
tv_grp_data= pd.read_csv(r'D:\Christ\Python\Assignment\tv_grp_data.csv')
video_impressions_data = pd.read_csv(r'D:\Christ\Python\Assignment\video_impressions_data.csv')
```

```python
# merge sales_data,tv_grp_data datasets on the 'Timestamp' column
merged_data=pd.merge(sales_data,tv_grp_data,on='Timestamp',how='outer')
merged_data
```

```python
# merge the merged_data with video_impressions_data on the 'Timestamp' column
final_data=pd.merge(merged_data,video_impressions_data,on='Timestamp',how='outer')
final_data
```
