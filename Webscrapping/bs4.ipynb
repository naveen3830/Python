```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.splashtop.com/"
response=requests.get(url)
response.content
```

```python
# The prettify() function in BeautifulSoup (BS4) is used to format the HTML content in a more readable and structured way.
soup=BeautifulSoup(response.content,'lxml')
soup.prettify()
```

```python
# find returns the first occurrence of a tag
heading=soup.find('h1')
heading
```

```python
# find_all returns the first occurrence of a tag
headings=soup.find_all('h2')
print(headings,'\n')

# The .text attribute is used to extract the text content from a HTML element, without any HTML tags.
for content in headings:
    print(content.text)
```

```python
for content in headings:
    print(content.get_text(strip=True))
```

```python
div_content = soup.find('div', class_='css-iylf1u-MobileMenu e9nvs208')
if div_content:
    print("Div Content:")
    print(div_content.text.strip())
```

```python
anchor = div_content.find('a', class_='css-wttjmx-NavHeader ek04rp13') if div_content else None
if anchor:
    print("\nAnchor Tag Content:")
    print(anchor.text.strip()) 
else:
    print("Anchor tag not found.")
```

```python
url='https://in.talent.com/jobs?k=&l=Bengaluru%2C+Karnataka&context=&s=&id=d2180ec14e2b'
response=requests.get(url).text
response
```

```python
soup=BeautifulSoup(response,'lxml')
```

```python
jobs=soup.find('section',class_='card card__job')
jobs
```

```python
company_name=jobs.find('div',class_="card__job-empname-label").text
company_name.replace(' ','')
```

```python
jobs=soup.find_all('section',class_='card card__job')

for job in jobs:
    post_name=job.find('h2',class_="card__job-title card__job-link gojob").text
    company_name=job.find('div',class_="card__job-empname-label").text
    posted_date=job.find('div',class_="c-card__jobDatePosted").text
    print(f"Company name: {company_name}")
    print(f"Post: {post_name} \n")
```
